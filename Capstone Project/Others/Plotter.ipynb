# Plotter

Notebook to plot graph of numeric feature distributions.

```python
# Import data
import pandas as pd
bank = pd.read_csv('bank-full.csv', sep=';')
```

```python
# Define numerical features
numerical = ['age', 'balance', 'campaign', 'pdays', 'previous']
```

```python
# Set matplotlib 
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# Plot
fig, axes = plt.subplots(nrows=3, ncols=2)
figsize = (12,12)

for i, feature in enumerate(numerical):
    minv = bank[feature].min()
    maxv = bank[feature].max()
    if i < 3:
        bank[feature].plot(kind='hist', figsize=figsize, ax=axes[i, 0], alpha=0.9, )
        axes[i, 0].set_title(str(feature).upper(), fontsize=12)
        axes[i, 0].set_xlabel(str(feature).capitalize() + ' Values' + '\n' + \
                             'Min: ' + str(minv) + '  Max: ' + str(maxv))
    else:
        bank[feature].plot(kind='hist', figsize=figsize, ax=axes[(i-3), 1], alpha=0.9)
        axes[(i-3), 1].set_title(str(feature).upper())
        axes[(i-3), 1].set_xlabel(str(feature).capitalize() + ' Values' + '\n' + \
                             'Min: ' + str(minv) + '  Max: ' + str(maxv))

fig.delaxes(axes[2, 1])
plt.tight_layout()
plt.show()
fig.savefig('feature_hist.pdf')
```
